## This Jupiter notebook contains the code that takes the data from mysql and feed into the trained model

pretrained models that are saved are used which are not uploaded due to the size

```python
import mysql.connector
from datetime import date
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from matplotlib import pyplot as plt
import joblib
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="",
  database="pirple"
)
```

```python
mycursor = mydb.cursor()

mycursor.execute("SHOW TABLES")

for x in mycursor:
  print(x)
```

```python
today = date.today()
sql = f"SELECT title,text FROM news WHERE date ='{today}'"
print(sql)

mycursor.execute(sql)

myresult = mycursor.fetchall()
```

```python
dataframe = pd.DataFrame(data=myresult,columns=['title','text'])
```

```python
dataframe
```

```python
if 'combined' not in dataframe.columns:
    dataframe.insert(len(dataframe.columns),'combined',None)

for row_index in dataframe.index:
    dataframe.loc[row_index,'combined'] = dataframe.loc[row_index,'title']+' '+dataframe.loc[row_index,'text']
```

```python
dataframe
```

```python
combined = pd.read_csv('combined.csv')
```

```python
v = TfidfVectorizer()
tfidf = v.fit(combined['combined'])
x = v.transform(dataframe['combined'])
```

```python
filename = './Model_Save/random_forest_1000.joblib' #choose which model
```

```python
loaded_model = joblib.load(filename)
result_1000 = loaded_model.predict(x)
```

```python
dataframe['results_rfr_1000']=result_1000
```

```python
filename = './Model_Save/random_forest_5000.joblib' #choose which model
```

```python
loaded_model = joblib.load(filename)
result_5000 = loaded_model.predict(x)
```

```python
dataframe['results_rfr_5000']=result_5000
```
